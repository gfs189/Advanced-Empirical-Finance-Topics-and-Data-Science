```python
#| label: ex2-functions

import quadprog

# ── Core portfolio functions (tidy-finance pattern) ───────────────────────────

def prefactor_sigma(Sigma, mu):
    """
    Solve Σ x = [μ | ι] in one LU factorisation and return the derived scalars.
    Following tidy-finance compute_efficient_weight: factorise Σ once per month,
    then each strategy's weights are O(N) vector operations.
    Returns (Si_mu, Si_iota, iota_Si_mu, iota_Si_iota).
    """
    N    = len(mu)
    iota = np.ones(N)
    try:
        sol = np.linalg.solve(Sigma, np.column_stack([mu, iota]))
    except np.linalg.LinAlgError:
        Si  = np.linalg.pinv(Sigma)
        sol = Si @ np.column_stack([mu, iota])
    Si_mu, Si_iota = sol[:, 0], sol[:, 1]
    return Si_mu, Si_iota, float(iota @ Si_mu), float(iota @ Si_iota)


def compute_efficient_weight(Si_mu, iota_Si_mu, Si_iota, iota_Si_iota,
                              gamma, lam=0.0, w_prev=None):
    """
    O(N) closed-form TC-adjusted MV using pre-factored Σ⁻¹μ and Σ⁻¹ι.

    Tidy-finance key identity: Σ⁻¹μ̃ = Σ⁻¹μ + 2λω⁺
    because μ̃ = μ + 2λΣω⁺  →  Σ⁻¹μ̃ = Σ⁻¹μ + 2λω⁺.
    So once Σ is factorised via prefactor_sigma(), each strategy costs O(N).
    """
    N = len(Si_mu)
    if lam == 0.0 or w_prev is None:
        w_prev, lam = np.zeros(N), 0.0
    gamma_tilde      = gamma + 2.0 * lam
    Si_mu_tilde      = Si_mu + 2.0 * lam * w_prev
    iota_Si_mu_tilde = iota_Si_mu + 2.0 * lam        # ι'ω⁺ = 1 always
    phi              = (iota_Si_mu_tilde - gamma_tilde) / iota_Si_iota
    return (1.0 / gamma_tilde) * (Si_mu_tilde - phi * Si_iota)


def compute_constrained_weights(mu, Sigma, gamma=GAMMA):
    """
    Long-only MV via quadprog — the Python equivalent of R's solve.QP from
    tidy-finance.org.  Falls back to cvxpy/OSQP then scipy SLSQP if needed.

    QP form:  min  ½ w'(γΣ)w − μ'w   s.t.  ι'w = 1,  w ≥ 0
    """
    N    = len(mu)
    iota = np.ones(N)
    w0   = iota / N

    # ── quadprog (R solve.QP equivalent, fastest) ─────────────────────────────
    try:
        G = (gamma * Sigma + 1e-8 * np.eye(N)).astype(np.float64)
        a = mu.astype(np.float64)
        C = np.column_stack([iota, np.eye(N)]).astype(np.float64)
        b = np.concatenate([[1.0], np.zeros(N)]).astype(np.float64)
        w = quadprog.solve_qp(G, a, C, b, meq=1)[0]
        w = np.maximum(w, 0.0); s = w.sum()
        return w / s if s > 1e-12 else w0
    except Exception:
        pass

    # ── cvxpy / OSQP fallback ────────────────────────────────────────────────
    try:
        import cvxpy as cp
        w    = cp.Variable(N)
        obj  = cp.Maximize(mu @ w - (gamma / 2.0) * cp.quad_form(w, cp.psd_wrap(Sigma)))
        prob = cp.Problem(obj, [cp.sum(w) == 1, w >= 0])
        prob.solve(solver=cp.OSQP, warm_start=True, eps_abs=1e-8, eps_rel=1e-8, verbose=False)
        if prob.status in ("optimal", "optimal_inaccurate") and w.value is not None:
            out = np.maximum(w.value, 0.0); s = out.sum()
            return out / s if s > 1e-12 else w0
    except Exception:
        pass

    # ── scipy SLSQP last resort ───────────────────────────────────────────────
    def neg_u(w): return -(w @ mu - (gamma / 2.0) * w @ Sigma @ w)
    def grad(w):  return -(mu - gamma * Sigma @ w)
    res = minimize(neg_u, w0, jac=grad, method="SLSQP",
                   bounds=[(0.0, None)] * N,
                   constraints=[{"type": "eq", "fun": lambda w: w.sum() - 1,
                                  "jac": lambda w: iota}],
                   options={"maxiter": 1000, "ftol": 1e-9})
    out = np.maximum(res.x if res.success else w0, 0.0)
    s   = out.sum()
    return out / s if s > 1e-12 else w0
```

```python
#| label: ex3-cov-functions

def estimate_factor_cov(R, F):
    """
    Factor model covariance Σ = B·ΣF·B' + diag(σ²_ε).
    Returns full N×N matrix — used only for the ex3 comparison plot.
    Backtest uses estimate_factor_components + woodbury_factor instead.
    """
    T, N = R.shape
    K    = F.shape[1]
    X_f  = np.column_stack([np.ones(T), F])
    coef = np.linalg.solve(X_f.T @ X_f, X_f.T @ R)
    betas     = coef[1:].T
    resid_var = np.var(R - X_f @ coef, axis=0, ddof=K + 1)
    Sigma_F   = np.cov(F.T, ddof=1)
    Sigma     = betas @ Sigma_F @ betas.T + np.diag(resid_var)
    lmin = np.linalg.eigvalsh(Sigma).min()
    if lmin < 1e-10:
        Sigma += (abs(lmin) + 1e-8) * np.eye(N)
    return Sigma


def estimate_factor_components(R, F):
    """Return (betas N×K, Sigma_F K×K, resid_var N) without forming Σ."""
    T, N = R.shape
    K    = F.shape[1]
    X_f  = np.column_stack([np.ones(T), F])
    coef = np.linalg.solve(X_f.T @ X_f, X_f.T @ R)
    betas     = coef[1:].T
    resid_var = np.maximum(np.var(R - X_f @ coef, axis=0, ddof=K + 1), 1e-8)
    return betas, np.cov(F.T, ddof=1), resid_var


def woodbury_factor(betas, Sigma_F, resid_var, V):
    """
    (B·ΣF·B' + D)⁻¹·V via Woodbury identity — O(NK²) instead of O(N³).
    Core system is K×K = 3×3. No N×N matrix formed or stored.
    D⁻¹v − D⁻¹B·(ΣF⁻¹ + B'D⁻¹B)⁻¹·B'D⁻¹v
    """
    D_inv   = 1.0 / resid_var
    D_inv_B = D_inv[:, None] * betas                         # N×K
    M       = np.linalg.inv(Sigma_F) + betas.T @ D_inv_B    # K×K  ← tiny
    D_inv_V = D_inv[:, None] * V                             # N×m
    return D_inv_V - D_inv_B @ np.linalg.solve(M, betas.T @ D_inv_V)


def lw_components(R_vals, lw_alpha):
    """
    Decompose Σ_LW = U'U + γI where U is T×N (T=60).
    U = sqrt((1−α)/(T−1)) · R_centred,  γ = α · trace(S)/N.
    Never forms the N×N matrix — caller uses woodbury_lw for solves.
    """
    R_c   = R_vals - R_vals.mean(axis=0)
    T     = R_c.shape[0]
    gamma = lw_alpha * np.var(R_c, axis=0, ddof=1).mean()
    U     = np.sqrt((1.0 - lw_alpha) / (T - 1)) * R_c       # T×N
    return U, gamma


def woodbury_lw(U, gamma, V):
    """
    (U'U + γI)⁻¹·V via Woodbury — O(T²N) instead of O(N³).
    Core system is T×T = 60×60.  Formula:
    V/γ − U'·(γI + UU')⁻¹·(UV)/γ
    """
    core = gamma * np.eye(U.shape[0]) + U @ U.T    # T×T  ← the only inversion
    return V / gamma - (U.T @ np.linalg.solve(core, U @ V)) / gamma


def simplex_project(v):
    """Project onto {w ≥ 0, Σw = 1} in O(N log N)."""
    u    = np.sort(v)[::-1]
    cssv = np.cumsum(u) - 1.0
    rho  = np.where(u > cssv / np.arange(1, len(v) + 1))[0][-1]
    return np.maximum(v - cssv[rho] / (rho + 1), 0.0)


def constrained_pg(mu, U, gamma, gamma_mv, n_iter=600, tol=1e-12):
    """
    Long-only MV via projected gradient + Woodbury Σ·w products.
    Each iteration O(TN) — no N×N QP, no N×N matrix.
    LW shrinkage keeps condition number low → fast convergence.
    """
    N    = U.shape[1]
    step = 1.0 / (gamma_mv * (np.linalg.norm(U, ord=2) ** 2 + gamma))
    w    = np.ones(N) / N
    for _ in range(n_iter):
        g     = gamma_mv * (U.T @ (U @ w) + gamma * w) - mu
        w_new = simplex_project(w - step * g)
        if np.dot(w_new - w, w_new - w) < tol:
            break
        w = w_new
    return w


def min_var_weights(Sigma):
    """Global minimum-variance portfolio (closed form) — used in ex3 plot."""
    iota = np.ones(Sigma.shape[0])
    try:
        SiI = np.linalg.solve(Sigma, iota)
    except np.linalg.LinAlgError:
        SiI = np.linalg.pinv(Sigma) @ iota
    return SiI / (iota @ SiI)
```

```python
#| label: ex3-comparison
#| fig-cap: Minimum-variance weights from the factor model ($\hat\Sigma$, x-axis) versus Ledoit-Wolf ($\hat\Sigma^{LW}$, y-axis) at December 2019. Each point is one stock; the dashed line is the 45° reference.

# Representative month for comparison (December 2019)
rep_date    = pd.Timestamp("2019-12-31")
win_dates   = returns_wide.index[returns_wide.index <= rep_date][-COV_WINDOW:]
R_rep       = returns_wide.loc[win_dates]

# Keep stocks with ≥ MIN_OBS valid observations, then drop stocks with any NaN
valid_mask  = R_rep.notna().sum() >= MIN_OBS
R_rep       = R_rep.loc[:, valid_mask].dropna(axis=1)

F_rep       = factors_ff3.reindex(R_rep.index).values
R_vals      = R_rep.values

Sigma_fac   = estimate_factor_cov(R_vals, F_rep)

lw_fitter   = LedoitWolf(assume_centered=False)
lw_fitter.fit(R_vals)
Sigma_lw_rep = lw_fitter.covariance_

w_fac = min_var_weights(Sigma_fac)
w_lw  = min_var_weights(Sigma_lw_rep)

cov_df = pd.DataFrame({"w_factor": w_fac, "w_lw": w_lw})
corr_w = np.corrcoef(w_fac, w_lw)[0, 1]
mae_w  = np.mean(np.abs(w_fac - w_lw))

(
    ggplot(cov_df, aes(x="w_factor", y="w_lw"))
    + geom_point(alpha=0.35, size=0.8)
    + geom_abline(intercept=0, slope=1, color="red", linetype="dashed")
    + theme_bw()
    + labs(title="Min-variance weights: factor model vs. Ledoit-Wolf (Dec 2019)",
           x="Factor model weight", y="Ledoit-Wolf weight")
).draw()

display(Markdown(
    f"The two estimators produce min-variance weights with correlation "
    f"**{corr_w:.3f}** and mean absolute difference **{mae_w*100:.2f} pp**. "
    f"{'The differences are small, suggesting both estimators agree on the broad structure of risk.' if mae_w < 0.005 else 'The non-trivial differences reflect the distinct regularisation philosophies of the two estimators.'}"
))
```

```python
#| label: ex4-predictions

formation_dates = pd.date_range(start=FORM_START, end=FORM_END, freq="ME")

# ── 1. Filter df_raw to all formation dates in one pass ───────────────────────
df_form = (
    df_raw
    .loc[df_raw["date"].isin(set(formation_dates)), ID_COLS + CHAR_COLS + MACRO_COLS]
    .copy()
    .sort_values(["date", "permno"])
    .reset_index(drop=True)
)

# ── 2. Imputation — one vectorised groupby instead of 180 per-month loops ─────
cs_med = df_form.groupby("date")[CHAR_COLS].transform("median")
df_form[CHAR_COLS] = df_form[CHAR_COLS].fillna(cs_med)

# Macros are date-level (same for every stock in a month): fill within month,
# then forward-fill any month that was entirely missing.
df_form[MACRO_COLS] = df_form.groupby("date")[MACRO_COLS].transform("first")
df_form[MACRO_COLS] = df_form[MACRO_COLS].ffill()
df_form = df_form.dropna(subset=MACRO_COLS).reset_index(drop=True)

# ── 3. SIC dummies ────────────────────────────────────────────────────────────
df_form["sic_division"] = sic2_to_division(df_form["sic2"])
sic_dummies = pd.get_dummies(
    pd.Categorical(df_form["sic_division"], categories=SIC_CATS_SORTED),
    prefix="sic", drop_first=True, dtype=float,
)

# ── 4. Interaction terms via numpy broadcasting — no Python loop ───────────────
# IXN_COLS uses product(CHAR_COLS, MACRO_COLS) — defined in setup, same order
# as the nested for-loop used in MA2 training so RF column names match exactly.
# (N, 10, 1) × (N, 1, 4) → (N, 10, 4) → reshape to (N, 40): all 40 products
# computed in a single numpy operation instead of 40 Python assignments.
chars_arr = df_form[CHAR_COLS].to_numpy()
macro_arr = df_form[MACRO_COLS].to_numpy()
ixn_arr   = (chars_arr[:, :, None] * macro_arr[:, None, :]).reshape(len(df_form), -1)

# ── 5. Assemble and predict in a single batch call ────────────────────────────
X_all = pd.concat([
    df_form[["permno", "date"] + CHAR_COLS].reset_index(drop=True),
    pd.DataFrame(ixn_arr, columns=IXN_COLS),
    sic_dummies.reset_index(drop=True),
], axis=1).fillna(0.0)

# One rf_model.predict call across all months — ~180× less Python overhead
# than the previous per-month loop.
mu_hat_all = rf_model.predict(X_all.drop(columns=["permno", "date"]))

mu_predictions = X_all[["permno", "date"]].assign(mu_hat=mu_hat_all)
```

```python
#| label: ex5-backtest

# ── Pre-computed lookups ───────────────────────────────────────────────────────
# Universe capped to top MAX_N_STOCKS by lagged market cap: reduces N from
# ~1000 to 300, cutting O(N³) linear-algebra cost by 37× before Woodbury.
univ_mktcap_dict = (
    universe_pq
    .sort_values("mktcap_lag", ascending=False)
    .groupby("date")
    .head(MAX_N_STOCKS)
    .groupby("date")["permno"]
    .apply(np.array)
    .to_dict()
)
rw_di      = returns_wide.index
rw_col_set = set(returns_wide.columns)

mu_dict = {
    date: grp.set_index("permno")["mu_hat"]
    for date, grp in mu_predictions.groupby("date")
}


def get_w_plus_vec(prev_w, prev_permnos, curr_permnos, drift):
    N = len(curr_permnos)
    if prev_w is None or prev_permnos is None:
        return np.ones(N) / N
    prev_ser = pd.Series(prev_w, index=prev_permnos)
    drifted  = prev_ser * drift.reindex(prev_permnos, fill_value=1.0)
    w_plus   = drifted.reindex(curr_permnos, fill_value=0.0).values
    total    = w_plus.sum()
    return w_plus / total if total > 1e-12 else np.ones(N) / N


# ── Backtest loop ─────────────────────────────────────────────────────────────
formation_dates = pd.date_range(start=FORM_START, end=FORM_END, freq="ME")

results      = []
prev_weights = {s: None for s in ["naive", "tc_mv", "tc_mv_lw", "constrained"]}
prev_permnos = None

for idx, t in enumerate(formation_dates):

    # ── 1. Universe: top MAX_N_STOCKS by mktcap (O(1) lookup) ────────────────
    permnos_t = univ_mktcap_dict.get(t, np.array([]))
    if len(permnos_t) < 10:
        continue

    # ── 2. Rolling 60-month window via binary search ──────────────────────────
    pos = int(rw_di.searchsorted(t, side="right"))
    if pos < MIN_OBS:
        continue
    win_dates = rw_di[max(0, pos - COV_WINDOW): pos]

    permnos_t = permnos_t[np.isin(permnos_t, list(rw_col_set))]
    if len(permnos_t) < 10:
        continue

    R_win     = returns_wide.loc[win_dates, permnos_t]
    valid     = R_win.notna().sum() >= MIN_OBS
    permnos_t = permnos_t[valid.values]
    R_win     = R_win[permnos_t].dropna(axis=1)
    permnos_t = R_win.columns.values

    if R_win.shape[0] < MIN_OBS or R_win.shape[1] < 10:
        continue

    N      = R_win.shape[1]
    R_vals = R_win.values

    # ── 3. Factor data ────────────────────────────────────────────────────────
    F_win = factors_ff3.reindex(R_win.index).values
    if np.isnan(F_win).any():
        continue

    # ── 4. Covariance — Woodbury components only, no N×N matrix ──────────────
    # Factor model Σ = B·ΣF·B' + D: store (betas, Sigma_F, resid_var).
    # woodbury_factor solves Σ⁻¹v in O(NK²) with K=3 — core is 3×3.
    betas_t, Sigma_F_t, resid_var_t = estimate_factor_components(R_vals, F_win)

    # Ledoit-Wolf Σ_LW = U'U + γI: store (U T×N, gamma scalar).
    # woodbury_lw solves Σ_LW⁻¹v in O(T²N) — core is T×T = 60×60.
    lw_fit       = LedoitWolf(assume_centered=False)
    lw_fit.fit(R_vals)
    U_t, gamma_t = lw_components(R_vals, lw_fit.shrinkage_)

    # ── 5. Expected returns ───────────────────────────────────────────────────
    mu_series    = mu_dict.get(t, pd.Series(dtype=float))
    mu_reindexed = mu_series.reindex(permnos_t)
    mu_t         = mu_reindexed.fillna(mu_reindexed.mean()).values

    # ── 6. Pre-rebalancing weights ────────────────────────────────────────────
    if pos > 0 and rw_di[pos - 1] == t:
        drift = 1.0 + returns_wide.iloc[pos - 1].fillna(0.0)
    else:
        drift = pd.Series(1.0, index=returns_wide.columns)

    wp_mv = get_w_plus_vec(prev_weights["tc_mv"],        prev_permnos, permnos_t, drift)
    wp_lw = get_w_plus_vec(prev_weights["tc_mv_lw"],     prev_permnos, permnos_t, drift)
    wp_c  = get_w_plus_vec(prev_weights["constrained"],   prev_permnos, permnos_t, drift)
    wp_n  = get_w_plus_vec(prev_weights["naive"],         prev_permnos, permnos_t, drift)

    # ── 7. Weights ────────────────────────────────────────────────────────────
    iota    = np.ones(N)
    V       = np.column_stack([mu_t, iota])
    w_naive = iota / N

    # Factor Σ: one Woodbury solve (3×3 core) → prefactors for TC-MV
    try:
        sol_t          = woodbury_factor(betas_t, Sigma_F_t, resid_var_t, V)
        Si_mu_t, Si_iota_t = sol_t[:, 0], sol_t[:, 1]
        w_tc_mv = compute_efficient_weight(
            Si_mu_t, float(Si_mu_t.sum()), Si_iota_t, float(Si_iota_t.sum()),
            GAMMA, LAMBDA_TC, wp_mv)
    except Exception:
        w_tc_mv = w_naive.copy()

    # LW Σ: one Woodbury solve (60×60 core) → prefactors for TC-MV-LW
    try:
        sol_lw          = woodbury_lw(U_t, gamma_t, V)
        Si_mu_lw, Si_iota_lw = sol_lw[:, 0], sol_lw[:, 1]
        w_tc_lw = compute_efficient_weight(
            Si_mu_lw, float(Si_mu_lw.sum()), Si_iota_lw, float(Si_iota_lw.sum()),
            GAMMA, LAMBDA_TC, wp_lw)
    except Exception:
        w_tc_lw = w_naive.copy()

    # Constrained: projected gradient — O(TN) per iter, no N×N QP at all
    try:
        w_con = constrained_pg(mu_t, U_t, gamma_t, GAMMA)
    except Exception:
        w_con = w_naive.copy()

    # ── 8. Next-month returns ─────────────────────────────────────────────────
    if pos >= len(rw_di):
        break
    t_next = rw_di[pos]
    r_next = returns_wide.loc[t_next, permnos_t].fillna(0.0).values

    # ── 9. TC and turnover — Woodbury quadratic forms, no N×N multiply ────────
    # Factor: d'Σd = (B'd)'ΣF(B'd) + resid_var·d²  →  O(NK²)
    # LW:     d'Σ_LW·d = ‖Ud‖² + γ‖d‖²             →  O(TN)
    def qtc_f(w, wp):
        d = w - wp; Bd = betas_t.T @ d
        return float(LAMBDA_TC * (Bd @ Sigma_F_t @ Bd + resid_var_t @ (d ** 2)))

    def qtc_lw(w, wp):
        d = w - wp; Ud = U_t @ d
        return float(LAMBDA_TC * (Ud @ Ud + gamma_t * (d @ d)))

    def to_(w, wp): return float(np.abs(w - wp).sum())

    results.append({
        "date":           t_next,
        "N":              N,
        "r_naive":        float(w_naive  @ r_next) - qtc_lw(w_naive,  wp_n),
        "r_tc_mv":        float(w_tc_mv  @ r_next) - qtc_f( w_tc_mv,  wp_mv),
        "r_tc_mv_lw":    float(w_tc_lw  @ r_next) - qtc_lw(w_tc_lw,  wp_lw),
        "r_constrained":  float(w_con    @ r_next) - qtc_lw(w_con,    wp_c),
        "to_naive":       to_(w_naive,  wp_n),
        "to_tc_mv":       to_(w_tc_mv,  wp_mv),
        "to_tc_mv_lw":   to_(w_tc_lw,  wp_lw),
        "to_constrained": to_(w_con,    wp_c),
    })

    prev_weights = {
        "naive": w_naive, "tc_mv": w_tc_mv,
        "tc_mv_lw": w_tc_lw, "constrained": w_con,
    }
    prev_permnos = permnos_t
    if (idx + 1) % 12 == 0:
        print(f"  Completed through {t.strftime('%Y-%m')} | N={N}")

bt = pd.DataFrame(results).set_index("date")
print(f"Backtest complete: {len(bt)} monthly observations.")
```
